# **Extracting 5 year TRAIN DATA [2019 - 2023]**

```python
import yfinance as yf
import pandas as pd

# Load the Excel file containing the stock tickers #downloaded from SET website
file_path = '/content/drive/MyDrive/Senior Project/Correct/DataPrep/MAI_List.xlsx'
df_tickers = pd.read_excel(file_path)

# All the stock tickers are in a column named 'Ticker'
tickers = df_tickers['Ticker'].tolist()

# Add '.BK' suffix to each ticker for Thai stocks b/c naming convention in yfinance
tickers = [ticker + '.BK' for ticker in tickers]

# Specify the date range
start_date = '2013-12-19'  # Start date (YYYY-MM-DD)
end_date = '2024-12-31'    # End date (YYYY-MM-DD)

# List to store tickers that have been listed for the full period
listed_tickers = []

# Loop through each ticker and check if the stock was listed for the full period
for ticker in tickers:
    stock = yf.Ticker(ticker)

    # Get historical prices for the specified period
    hist = stock.history(period="max")  # Get all available data

    # Check if data exists and spans the full date range
    if not hist.empty:
        # Ensure data is available for the full period and remove any timezone information
        hist.index = hist.index.tz_localize(None)
        data_start = hist.index.min().date()  # Convert to date for comparison
        data_end = hist.index.max().date()    # Convert to date for comparison

        # Check if the stock was listed on or before 2014-01-01 and is still listed on or after 2022-12-31
        if data_start <= pd.to_datetime(start_date).date() and data_end >= pd.to_datetime(end_date).date():
            listed_tickers.append(ticker)
        else:
            print(f"Ticker {ticker} was not listed for the full period from {start_date} to {end_date}, skipping...")

    else:
        print(f"No data found for {ticker}, skipping...")

# Print and save the tickers with full listing data
print(f"Tickers listed from {start_date} to {end_date}:")
for listed_ticker in listed_tickers:
    print(listed_ticker)

# Optionally, save listed tickers to a file
output_file = '/content/drive/MyDrive/Senior Project/Correct/DataPrep/Selected_stocksFINAL.xlsx'
pd.DataFrame(listed_tickers, columns=['Ticker']).to_excel(output_file, index=False)

print(f"Tickers listed for the full period saved to {output_file}")
```

# **Lets Prepare the data for ANN input features**

# **Train Data First**

```python
!pip install XlsxWriter
# -------- SETTINGS --------
#  path to Google Drive)
input_path = '/content/drive/My Drive/Senior Project/Correct/2nd/DataPrep/Train5YearsFinal.xlsx'
output_path = '/content/drive/My Drive/Senior Project/Correct/2nd/features_for_ANN/features_for_ANN_Train.xlsx'

# -------- FUNCTION TO CREATE FEATURES --------
def prepare_custom_features(df):
    df = df.sort_values(by='Date')  # ensure chronological order

    df['todayOpen'] = df['Open']
    df['HighPreviousday'] = df['High'].shift(1)
    df['LowPreviousday'] = df['Low'].shift(1)
    df['Vol_Previousday'] = df['Volume'].shift(1)
    df['prev7dayAvgClose'] = df['Close'].shift(1).rolling(window=7).mean()
    df['Price'] = df['Close']  # Today's closing price

    df = df[['Date', 'todayOpen', 'HighPreviousday', 'LowPreviousday',
             'Vol_Previousday', 'prev7dayAvgClose', 'Price']]

    df.dropna(inplace=True)
    return df

# -------- MAIN PROCESS --------
def main():
    xls = pd.ExcelFile(input_path)
    sheet_names = xls.sheet_names

    with pd.ExcelWriter(output_path, engine='xlsxwriter') as writer:
        for sheet in sheet_names:
            df = xls.parse(sheet)
            required_cols = {'Date', 'Open', 'High', 'Low', 'Close', 'Volume'}
            if required_cols.issubset(df.columns):
                df_features = prepare_custom_features(df)
                df_features.to_excel(writer, sheet_name=sheet[:31], index=False)
                print(f"✅ Processed: {sheet}")
            else:
                print(f"❌ Skipped {sheet}: missing columns")

    print(f"\n✅ All sheets processed and saved to {output_path}")

# -------- RUN --------
if __name__ == "__main__":
    main()
```

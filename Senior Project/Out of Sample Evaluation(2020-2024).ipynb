```python
import pandas as pd
import yfinance as yf

# --- Step 1: Load top 20 tickers from expected returns file ---
top_20_df = pd.read_csv('/content/drive/My Drive/Senior Project/Correct/Input_for_models/top_20_expected_returns.csv', header=None)
top_20_df = top_20_df.iloc[:, [0]]  # Keep only the first column
top_20_df.columns = ['Ticker']

# --- Step 2: Clean ticker list and add .BK suffix ---
listed_tickers = [
    str(ticker).strip() + ".BK" if ".BK" not in str(ticker) else str(ticker).strip()
    for ticker in top_20_df['Ticker']
    if pd.notna(ticker)
]

print("‚úÖ Tickers to download:", listed_tickers)
print("üìå Total tickers:", len(listed_tickers))

# --- Step 3: Set date range ---
start_date = '2020-01-01'
end_date = '2024-12-31'

# --- Step 4: Excel writer config ---
output_file_path = '/content/drive/My Drive/Senior Project/Correct/Evaluation/Selectedstocks_actualPriceData_2020_2024.xlsx'
writer = pd.ExcelWriter(output_file_path, engine='openpyxl')

# --- Step 5: Download and save to Excel ---
for ticker in listed_tickers:
    try:
        stock = yf.Ticker(ticker)
        hist = stock.history(start=start_date, end=end_date)

        if hist.empty:
            print(f"‚ö†Ô∏è No data found for {ticker}, skipping...")
            continue

        hist.index = hist.index.tz_localize(None)
        hist = hist.reset_index()[['Date', 'Close']]
        hist = hist.rename(columns={'Close': 'Price'})

        sheet_name = ticker
        hist.to_excel(writer, sheet_name=sheet_name, index=False)
        print(f"‚úÖ Saved: {sheet_name}")

    except Exception as e:
        print(f"‚ùå Error downloading {ticker}: {e}")

# --- Step 6: Save workbook ---
writer.close()
print(f"\n‚úÖ All data saved to: {output_file_path}")
```
